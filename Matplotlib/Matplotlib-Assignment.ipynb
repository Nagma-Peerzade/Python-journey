```python
'''1. Write a Python programming to display a bar chart of the popularity of
programming Languages. Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7'''
```

```python
import matplotlib.pyplot as plt
```

```python
Plang=['Java', 'Python', 'PHP', 'JavaScript',' C#',' C++']
Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(Plang,Popularity,color='pink',width=0.3)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.show()
```

```python
#2. Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.
Plang=['Java', 'Python', 'PHP', 'JavaScript',' C#',' C++']
Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.barh(Plang,Popularity,color='pink')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.show()
```

```python
#3. Write a Python programming to display a bar chart of the popularity of programming Languages. Use uniform color.
Plang=['Java', 'Python', 'PHP', 'JavaScript',' C#',' C++']
Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(Plang,Popularity,color='yellow',width=0.3,)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.show()
```

```python
#4 Write a Python programming to display a bar chart of the popularity of programming Languages. Use different color for each bar
plang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(plang, popularity, color=['skyblue', 'indigo', 'purple', 'pink', 'yellow', 'cyan'])
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
#5. Write a Python programming to display a bar chart of the popularity of programming Languages. Make blue border to each bar
plang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(plang, popularity,color='pink',ec='blue',width=0.3)
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
'''6.Write a Python program to create a stacked bar plot .
Note: Use bottom to stack the women bars on top of the men bars.
Sample Data: 
men = [122, 230,1 35,335, 126]
women = [225,132, 30, 235,229]'''
men = [122, 230,135,335, 126]
women = [225,132, 30, 235,229]  
plt.bar(range(len(men)),men,label='men',width=0.3,color='blue')
plt.bar(range(len(women)),women,bottom=men,label='women',width=0.3,color='pink')
plt.show()
```

```python
'''7. Write a Python programming to create a pie chart of the popularity of
programming Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7'''

plang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.pie(popularity,labels=plang)
plt.show()
```
