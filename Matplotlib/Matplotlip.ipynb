```python
name=['riya','siya','tiya','priya']
tmarks=[689,590,870,670]
plt.bar(name,tmarks,width=0.3,color='pink')
plt.xlabel('student names')
plt.ylabel('Marks')
plt.title('Result')
plt.show()
```

```python
#stacked bar plot
plt.figure(figsize=(8,4))
eng=[90,78,68,70]
sci=[90,67,89,56]
maths=[45,90,56,10]
bot3=[x+y for x,y in zip(eng,sci)]
name=['riya','siya','tiya','priya']
plt.bar(name,eng,width=0.4,label='Eng')
plt.bar(name,sci,bottom=eng,width=0.4,label='Sci')
plt.bar(name,maths,bottom=bot3,width=0.4,label='Maths')
plt.legend()
plt.show()
```

```python
#multiple

plt.figure(figsize=(8,4))
eng=[90,78,68,70]
sci=[90,67,89,56]
maths=[45,90,56,10]
name=['riya','siya','tiya','priya']
bar1=[2,4,6,8]
bar2=[i+0.4 for i in bar1]
bar3=[i+0.4 for i in bar2]
plt.bar(bar1,eng,width=0.3,label='Eng')
plt.bar(bar2,sci,width=0.3,label='Sci')
plt.bar(bar3,maths,width=0.3,label='Maths')
plt.legend()
plt.xlabel('Student Names')
plt.ylabel('Marks Scored')
plt.title('Result')
plt.xticks(bar2,name)
plt.show()
```

```python
#min,25 percntile,50 percentile,75 percentile
x=[1,2,3,4,5,6,7]
#x=[1,2,3,4,5,6,7,16,17]#outlier
plt.boxplot(x)
plt.show()
```

```python
#histogram
marks=[10,15,18,30,49,50,56,89]
plt.hist(marks,ec='red',bins=[0,11,21,31,41,51,61,71,81,91])
plt.xlabel('marks range')
plt.ylabel('count')
plt.show()
```

```python
#pie chart
subj=['eng','maths','sci','hindi']
stud1=[75,98,99,56]
plt.pie(stud1,labels=subj,autopct='%f%%')
plt.show()
```

```python
dnames=['satara','pune','mumbai','nashik']
pofE=[60,100,200,90]
pofW=[30,40,50,70]
pofS=[89,90,45,67]
pofN=[56,67,78,89]
plt.plot(dnames,pofE,'g:',label='East')
plt.plot(dnames,pofW,'b--',label='West')
plt.plot(dnames,pofS,'r:',label='South')
plt.plot(dnames,pofN,'y*',label='North')
plt.xlabel('District names')
plt.ylabel('Population')
plt.title('survey')
plt.legend(bbox_to_anchor=(0.75,1.30),ncol=2)
plt.show()
```

```python
#bar 
dnames=['satara','pune','mumbai','nashik']
pofE=[60,100,200,90]
pofW=[90,40,50,70]
pofS=[89,90,45,67]
pofN=[70,67,78,89]
bar1=[2,4,6,8]
bar2=[i+0.3 for i in bar1]
bar3=[i+0.3 for i in bar2]
bar4=[i+0.3 for i in bar3]
plt.bar(bar1,pofE,label='East',width=0.3,color='pink')
plt.bar(bar2,pofW,label='West',width=0.3,color='yellow')
plt.bar(bar3,pofS,label='South',width=0.3,color='skyblue')
plt.bar(bar4,pofN,label='North',width=0.3,color='cyan')
plt.xlabel('District names')
plt.ylabel('Population')
plt.legend()
plt.xticks(bar2,dnames)
plt.show()
```

```python
city=['Bgm','Pune','Mumbai','Chennai']
day1=[21.9,22.8,30.6,20.5]
day2=[12,17,23,24]
day3=[24,23,27,19]
bar1=[2,4,6,8]
bar2=[i+0.4 for i in bar1]
bar3=[i+0.4 for i in bar2]
plt.bar(bar1,day1,label='Temp_day1',color='yellow',width=0.4)
plt.bar(bar2,day2,label='Temp_day2',color='pink',width=0.4)
plt.bar(bar3,day3,label='Temp_day3',color='skyblue',width=0.4)
plt.xlabel('Cities')
plt.ylabel('Temprature')
plt.title('results')
plt.legend(bbox_to_anchor=(1.0,1.30),ncol=3)
plt.xticks(bar2,city)
plt.show()
```

```python
#stacked bar plot
dnames=['satara','pune','mumbai','nashik']
pofE=[60,100,200,90]
pofW=[90,40,50,70]
pofS=[45,89,90,67]
pofN=[78,60,45,60]
bot3=[x+y for x,y in zip(pofE,pofW)]
bot4=[x+y for x,y in zip(bot3,pofS)]
plt.bar(dnames,pofE,color='yellow',width=0.3,label='East')
plt.bar(dnames,pofW,bottom=pofE,color='cyan',width=0.3,label='West')
plt.bar(dnames,pofS,bottom=bot3,color='pink',width=0.3,label='South')
plt.bar(dnames,pofN,bottom=bot4,color='blue',width=0.3,label='North')
plt.xlabel('DIstrict names')
plt.ylabel('Population')
plt.title('Survey')
plt.legend(bbox_to_anchor=(0.75,1.30),ncol=2)
plt.show()
```
